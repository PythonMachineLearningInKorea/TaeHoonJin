```
column_number = {}
for i, column in enumerate(sample_submission.columns):
    column_number[column] = i
    
def to_number(x, dic):
    return dic[x]

train['type_num'] = train['type'].apply(lambda x: to_number(x, column_number))
```

```
train
```

```
train_x = train.drop(columns=['id','type', 'type_num'], axis=1)
train_y = train['type_num']
test_x = test
train_x
```

```
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(train_x, train_y, \
                                                  test_size=0.2, random_state=42, stratify = train_y)
```

```
%%time
xgb_clf = XGBClassifier(n_estimators=100, n_jobs=-1, random_state=42)
xgb_clf.fit(X_train,y_train, eval_set=[(X_train, y_train), (X_valid, y_valid)])
```

```
test_id=test['id']
test=test.drop(columns='id')
```

```
test_id
```

```
xgb_pred = xgb_clf.predict_proba(test)
```

```
xgb_pred
```

```
submission=pd.DataFrame(data=xgb_pred)
submission=pd.concat([test_id,submission],axis=1)
submission.columns=sample_submission.columns
submission
```
